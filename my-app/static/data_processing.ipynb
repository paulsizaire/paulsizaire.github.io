```python
#import libraries
import pandas as pd
import numpy as np
```

```python
socio_ec = pd.read_csv('socioec_ECF.csv')
socio_ec.columns
```

```python
# make 3 columns with top 1, 2, 3 places that migrants are from 
# mig_cols = ['MIG_AS_PERCENT',
#        'MIG_AS_E_PERCENT', 'MIG_AS_SC_PERCENT', 'MIG_AS_SE_PERCENT',
#        'MIG_AS_W_PERCENT', 'MIG_AFR_PERCENT', 'MIG_OC_PERCENT',
#        'MIG_LA_PERCENT', 'MIG_LA_CAR_PERCENT', 'MIG_LA_CAM_PERCENT',
#        'MIG_LA_CAM_MX_PERCENT', 'MIG_LA_SAM_PERCENT', 'MIG_NA_PERCENT',
#        'MIG_LA_CAM_OTHER_PERCENT']

mig_cols = [
       'MIG_AS_E_PERCENT', 'MIG_AS_SC_PERCENT', 'MIG_AS_SE_PERCENT',
       'MIG_AS_W_PERCENT', 'MIG_AFR_PERCENT', 'MIG_OC_PERCENT',
     'MIG_LA_CAR_PERCENT', 'MIG_LA_CAM_PERCENT',
       'MIG_LA_CAM_MX_PERCENT', 'MIG_LA_SAM_PERCENT', 'MIG_NA_PERCENT',
       'MIG_LA_CAM_OTHER_PERCENT']
```

```python
#top 1,2,3
top_1 = [socio_ec.loc[i,mig_cols].sort_values(ascending=False)[0:1].index[0] for i in np.arange(len(socio_ec))]
top_2 = [socio_ec.loc[i,mig_cols].sort_values(ascending=False)[1:2].index[0] for i in np.arange(len(socio_ec))]
top_3 = [socio_ec.loc[i,mig_cols].sort_values(ascending=False)[2:3].index[0] for i in np.arange(len(socio_ec))]
```

```python
#apply map
mig_dict = {
       'MIG_AS_E_PERCENT':'East Asia', 
       'MIG_AS_SC_PERCENT': 'South Central Asia', 
       'MIG_AS_SE_PERCENT': 'South East Asia',
       'MIG_AS_W_PERCENT': 'West Asia / Middle East', 
       'MIG_AFR_PERCENT': 'Africa', 
       'MIG_OC_PERCENT': 'Oceana',
        'MIG_LA_CAR_PERCENT': 'Caribbean' , 
        'MIG_LA_CAM_PERCENT': 'Central America',
       'MIG_LA_CAM_MX_PERCENT': 'Mexico', 
       'MIG_LA_SAM_PERCENT': 'South America', 
       'MIG_NA_PERCENT': 'North America',
       'MIG_LA_CAM_OTHER_PERCENT': 'Other Central American Region'}
```

```python
pd.Series(top_1).map(mig_dict)
```

```python
socio_ec['top_1']=pd.Series(top_1).map(mig_dict)
socio_ec['top_2']=pd.Series(top_2).map(mig_dict)
socio_ec['top_3']=pd.Series(top_3).map(mig_dict)
```
