```python
import pandas as pd
import requests
import io
    
# Downloading the csv file from what Kailin uploaded

url = "https://raw.githubusercontent.com/paulsizaire/paulsizaire.github.io/main/socioec_ECF.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

df = pd.read_csv(io.StringIO(download.decode('utf-8')))
```

```python
#looking at it 
df.FIPS.unique()
```

```python
# take only the columns with "TONCO2E"
emissions_cols = [col for col in df.columns if 'TONCO2E_' in col]
emissions_cols.append("FIPS")
df[emissions_cols]
```

```python
df[emissions_cols]
```

```python
df["FIPS"]=df["FIPS"].astype(int).astype(str)
```

```python
emissions_df = df[emissions_cols].T.reset_index().rename({'index':'industry'},axis=1)
#
emissions_df
```

```python
emissions_df.columns=emissions_df.iloc[8]
emissions_df.drop(8,inplace=True)
```

```python
emissions_df["FIPS"] = ['Agriculture','Construction','Commercial','Manufacturing','Mining','Power','Oil and Gas','Coal']
emissions_df
```

```python
emissions_df.to_csv("ACS/emissions.csv")
```
