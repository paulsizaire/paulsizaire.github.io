```python
import pandas as pd
import requests
import io
    
# Downloading the csv file from what Kailin uploaded

url = "https://raw.githubusercontent.com/paulsizaire/paulsizaire.github.io/socioeconomic-panel/ACS/socioec_ECF.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

df = pd.read_csv(io.StringIO(download.decode('utf-8')))
```

```python
#looking at it 
len(df.FIPS.unique())
```

```python
df["FIPS"] = df.FIPS.astype(int)
df.to_csv('ACS/socioec_ECF.csv')
```

## Make ECF df

```python
# ecf_df = df.loc[:,["ECF_ST","ECF_ST_log10","ECF_US","ECF_US_log10","ECF","ECF_log10","FIPS"]]
# ecf_df
ecf_cols = ["ECF","ECF_ST","ECF_US","FIPS"]
df["FIPS"]=df["FIPS"].astype(int).astype(str)
ecf_df = df.loc[:,ecf_cols]
ecf_df
```

```python
ecf_df = df[ecf_cols].rename({"ECF_ST":'State','ECF_US':'National','ECF':'County'},axis=1).T.reset_index().rename({'index':'scale'},axis=1).fillna(0)
ecf_df
```

```python
ecf_df.columns=ecf_df.iloc[3]
ecf_df.drop(3,inplace=True)
```

```python
ecf_df
```

```python
ecf_df.iloc[:,1:] = ecf_df.iloc[:,1:].astype(int)
```

```python
ecf_df
```

```python
ecf_df.to_csv('ACS/ecf.csv')
```

## Make employment df

```python
employment_cols = [col for col in df.columns if ('EMP_' in col) & ('PERCENT' in col)]
employment_cols.append("FIPS")
df["FIPS"]=df["FIPS"].astype(int).astype(str)
employment_df = df[employment_cols].fillna(0)

employment_df
```

```python
employment_df = df[employment_cols].T.reset_index().rename({'index':'industry'},axis=1).fillna(0)
```

```python
employment_df
```

```python
employment_df.columns=employment_df.iloc[8]
employment_df.drop(8,inplace=True)
```

```python
#check these
employment_df["FIPS"] = ['Agriculture','Construction','Commercial','Manufacturing','Mining','Power','Oil and Gas','Coal']
employment_df
```

```python
employment_df.to_csv("ACS/employment.csv")
```

## Make emissions df

```python
# take only the columns with "TONCO2E"
emissions_cols = [col for col in df.columns if 'TONCO2E_' in col]
emissions_cols.append("FIPS")
df[emissions_cols]
```

```python
df["FIPS"]=df["FIPS"].astype(int).astype(str)
```

```python
emissions_df = df[emissions_cols].T.reset_index().rename({'index':'industry'},axis=1).fillna(0)
emissions_df
```

```python
emissions_df.columns=emissions_df.iloc[8]
emissions_df.drop(8,inplace=True)
```

```python
#check these
emissions_df["FIPS"] = ['Agriculture','Construction','Commercial','Manufacturing','Mining','Power','Oil and Gas','Coal']
emissions_df
```
